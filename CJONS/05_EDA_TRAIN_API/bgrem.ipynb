### 배경 제거하기

```python
from rembg.bg import remove
```

```python
def show_rembg(i_path, o_path):
    
    #fig = plt.figure(figsize=(10, 10))
    
    #fig.add_subplot(1,2,1)
    orig_img = Image.open(i_path)
    #plt.imshow(orig_img)
    
    output = remove(orig_img)
    #fig.add_subplot(1,2,2)
    #plt.imshow(output)
    output.save(o_path)
```

```python
input_path = df.iloc[0]['file_loc']
file_path = list(input_path.split('/'))
print(file_path)

output_path = file_path[0] + '/' + file_path[1] + '/' + 'bgrem_' + file_path[2]
print(output_path)
```

```python
input_paths = []
output_paths = []


for i in range(len(df)):
    input_path = str(df.iloc[i]['file_loc'])
    input_paths.append(input_path)
    
    file_path = list(input_path.split('/'))
    
    output_path = str(file_path[0] + '/' + file_path[1] + '/' + 'bgrem_' + file_path[2])
    output_paths.append(output_path)
```

```python
for n in range(len(input_paths)):
    show_rembg(input_paths[n], output_paths[n])
```
